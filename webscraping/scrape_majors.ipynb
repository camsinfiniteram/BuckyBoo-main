```python
guide_url = "https://guide.wisc.edu/undergraduate/#majorscertificatestext"
driver = webdriver.Chrome()
driver.get(guide_url)
time.sleep(1)

driver = driver.find_element(By.ID, "majorscertificatestextcontainer")
driver = driver.find_element(By.CLASS_NAME, "clearfix")
links = driver.find_elements(By.XPATH, "//a")

hrefs_and_titles = []
for link in links:
    href = link.get_attribute("href")
    title = link.text  # Or use appropriate logic to extract the title
    hrefs_and_titles.append((href, title))
```

```python
# Filter out the links that are not to major pages
filtered_links = []
for href, title in hrefs_and_titles:
    if href and title and href.startswith("https://guide.wisc.edu/undergraduate/"):
        filtered_links.append((href, title))

major_links = filtered_links[1:]  # First one is garbage
major_links = major_links[:-14]  # last few are also garbage
major_links

# Append requrements to url
major_links = [(link[0] + "/#requirementstext", link[1]) for link in major_links]
```

```python
url = "https://guide.wisc.edu/undergraduate/letters-science/astronomy/astronomy-physics-ba/#requirementstext"
driver = webdriver.Chrome()

major_requirements = {}
for url, title in major_links:
    print(url, title)
    try:
        driver.get(url)

        requirements_tab = driver.find_element(
            By.CSS_SELECTOR, "a[href='#requirementstextcontainer']"
        )
        requirements_tab.click()
        time.sleep(1)

        requirements = driver.find_element(By.CSS_SELECTOR, "#requirementstextcontainer")

        # Extract text from the container
        # requirements_text_container = driver.find_element(By.ID, )
        requirements_text = requirements.get_attribute("innerText")

        major_requirements[title] = requirements_text
    except Exception as e:
        print(e)
```

```python
with open("major_requirements.json", "w") as file:
    file.write(json.dumps(major_requirements))
```
